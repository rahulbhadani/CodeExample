```python
import glob
import os
import numpy as np
from subprocess import call
import re
import subprocess
```

```python
def rename_files1(files):
    folder_name = files[0][0:files[0].find('/')]
    for e in files:
        search1 = "-SignalPos_"
        p = re.search(search1, e).start()
        e1 = "{}{}0{}".format(e[0:p], e[p:p + len(search1)], e[p + len(search1):])
        search2 = "-SignalLen_"
        q = re.search(search2, e1).start()
        e2 = "{}{}0{}".format(e1[0:q], e1[q:q + len(search2)], e1[q + len(search2):])
        print(e2)
        os.rename(e, e2)
```
