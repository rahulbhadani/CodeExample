```python
import inspect
def whoami():
    return inspect.stack()[1][3]

class kw(object):
    def __init__(self, testval, **kwargs):
        print("Testval :{}".format(testval))
        
        print("kwargs: {}".format(kwargs))
        
        print("kwargs keys: {}".format(kwargs.keys()))
        
        str = []
        for key in kwargs.keys():
            
            str.append("{}:={}".format(key, kwargs[key]))
        print(str)
    
    def thisclass(self):
        print(whoami())
```
